```python
# threshold between 0 and 1 
def checkFeatureForBalance(col, threshold):
    if col.dtypes != object:
        return print('Wrong type')
    numberOfrows = col.count()
    countedValues = col.value_counts()
    for val in countedValues: 
        if val > numberOfrows * threshold:
            return print('Imbalanced Column')
        else:
            return print('Balanced Column')
```

```python
checkFeatureForBalance(train_df.land_surface_condition, 0.9)
```
